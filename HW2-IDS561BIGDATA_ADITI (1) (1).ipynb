# STEP 3 calculate the sum of “favorite_count” and “retweet_count” for each tweet on that day

```python
#sum the favorite_count and retweet_count column

sum_column['sum_count'] = sum_column['favorite_count'] + sum_column['retweet_count']
```

```python
sum_column
```

```python
#getting the highest sum count

total_sum = sum_column.sort_values(by = 'sum_count', ascending=False)
```

# Calculating the top 100 tweets

```python
#checking the 'text_' column

data = total_sum.nlargest(100, 'sum_count')
```

# Count the word frequency of the 100 tweets

```python
def create_session():
  spk = SparkSession.builder \
      .appName("HW2") \
      .getOrCreate()
  return spk
spark = create_session()

import findspark
findspark.init()
```

```python
ss = SparkContext.getOrCreate()
```

```python
col_list =  list(datatext["text_"])
print(col_list)
```

```python
a = ss.parallelize([col_list])
```

```python
sp_df = a.flatMap(lambda line: line.split(" "))
```

```python
wordcount = sp_df.map(lambda word: (word, 1)).reduceByKey(lambda a,b:a +b)
```
